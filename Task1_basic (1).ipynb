```python
# загрузка nltk

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
```

```python
# загрузка spacy

import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# загрузка текста

!wget -O data.txt https://raw.githubusercontent.com/vifirsanova/compling/refs/heads/main/tasks/task1/data.txt
```

```python
# открытие текста

with open("data.txt", "r", encoding="utf-8") as file:
    data = file.read()
```

```python
# очистка текста от пунктуации с помошью регулярных выражений

import re

cleaned_text = re.sub(pattern = r'[\W]', repl = ' ', string = data) #с помощью \W заменила на пробелы всё то, что не является словами и цифрами, чтобы избавиться от пунктуации; замена на пробелы сделана, чтобы не потерять пробелы между словами (с \S не сработало)
print("Текст без пунктуации:", cleaned_text)
```

```python
# приведение текста к нижнему регистру

cleaned_text = word_tokenize(cleaned_text.lower())
print("Текст в нижнем регистре:", cleaned_text)
```

```python
# удаление стоп-слов с помощью nltk

from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in cleaned_text if word not in stop_words]
print("Текст без стоп-слов:", filtered_words)
```

```python
# превращение списка в строку

words=' '.join(filtered_words)
```

```python
# лемматизация текста с помощью spacy
lemmatized_words = nlp(words)
print("Лемматизированный текст:", lemmatized_words)
```
